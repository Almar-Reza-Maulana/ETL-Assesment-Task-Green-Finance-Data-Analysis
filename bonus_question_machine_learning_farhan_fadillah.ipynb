### Definisi dan Posisi Kuartil dalam Distribusi Data
Kuartil pertama (Q1) adalah nilai di bawahnya terdapat 25 data, sedangkan kuartil ketiga (Q3) adalah nilai di bawahnya terdapat 75 data.
Rentang antar kuartil (IQR = Q3 - Q1) menggambarkan sebaran tengah data yang merepresentasikan variabilitas utama data tanpa dipengaruhi oleh nilai ekstrim (outlier).
Oleh karena itu, Q1 dan Q3 adalah indikator lokal yang kuat untuk menangkap distribusi "normal" dari data.

### Mengapa Menggunakan Q1 dan Q3 untuk Synthetic Data Generation?
Ketika membuat data sintetis, tujuan utama adalah menghasilkan data yang representatif terhadap distribusi data asli, serta menghindari nilai ultra-ekstrem yang bisa mengganggu analisis atau model.
Nilai minimum dan maksimum absolut pada data nyata sering mengandung outlier yang tidak diinginkan.
Dengan menetapkan Q1 sebagai batas bawah minimum dan Q3 sebagai batas atas maksimum, synthetic data dapat dihasilkan dalam rentang yang masuk akal (IQR).
Pendekatan ini menjaga keaslian pola data (yakni distribusi tengah), sekaligus meminimalkan risiko noise berlebihan dari outlier.

### Pendekatan Statistik Lain yang Mendukung
Menggunakan IQR sebagai batas merupakan praktek standar dalam statistik untuk mendeteksi dan mengeliminasi outlier (contoh: aturan Tukey).
Dengan menggunakan Q1 dan Q3 untuk membatasi nilai sintetis, kita melakukan regularisasi data dan menjaga distribusi yang robust.
Pendekatan ini juga membantu dalam menjaga nilai-nilai data sintetis agar tetap biologis, finansial, atau secara domain-related valid.

### Implikasi Praktis untuk Synthetic Data
Synthetic data yang dihasilkan akan memiliki rentang nilai yang realistis dan representatif.
Algoritma pembelajaran yang menggunakan synthetic data ini akan lebih stabil dan akurat.
Mencegah overfitting akibat nilai ekstrem yang tidak representatif.
Mempermudah validasi dan interpretasi hasil analisis atau model.

### Landasan Teori Statistik: Kuartil dan Interquartile Range (IQR)
Kuartil adalah nilai yang membagi data menjadi empat bagian sama besar berdasarkan distribusi kumulatifnya. Q1 (kuartil pertama) menandai angka pada persentil ke-25, sementara Q3 (kuartil ketiga) menandai persentil ke-75.

Interquartile Range (IQR) adalah rentang antara Q3 dan Q1, yang meliputi 50 data tengah (data pusat). IQR adalah statistik robust, stabil terhadap outlier, dan menggambarkan sebaran atau variabilitas data utama secara efektif.

Dari sudut pandang teori probabilitas, IQR memberikan informasi tentang distribusi pusat (central tendency) dan dispersion (penyebaran) secara simultan, yang lebih informatif dibandingkan hanya rentang (range) karena rentang sangat dipengaruhi oleh outlier.

### Deteksi dan Penanganan Outlier
Secara standar dalam statistik (dikenal sebagai aturan Tukey), data yang berada di luar rentang [Q1 - 1.5IQR, Q3 + 1.5IQR] sering dianggap sebagai outlier atau nilai ekstrim yang tidak merepresentasikan populasi secara umum.

Dalam konteks pembuatan data sintetis, nilai-nilai ekstrem ini bisa menimbulkan bias dan gangguan model jika disertakan secara utuh.

Dengan menggunakan nilai Q1 sebagai batas bawah dan Q3 sebagai batas atas dalam pengelolaan data sintetis, kita mengeliminasi pengaruh outlier yang mungkin tidak diinginkan, sehingga data sintetis menjadi lebih stabil dan konsisten secara statistik.

### Pembuatan Data Sintetis: Preservasi Distribusi Asli
Synthetic data generation berupaya membuat data baru yang secara statistik serupa dengan data asli, mempertahankan karakteristik distribusi, kecenderungan sentral, dan hubungan antar variabel. Hal ini krusial agar data sintetis dapat digunakan secara aman untuk analisis lanjutan, pelatihan model, atau privasi data.
Salah satu pendekatan yang disarankan dalam literatur ilmiah adalah mempertahankan statistik ringkasan utama termasuk median, kuartil, dan IQR agar data sintetis sesuai dengan pola distribusi asli tetapi tidak mereplikasi noise dan ekstrim yang mengganggu.
Dengan membatasi ruang nilai (value range) sintetis sesuai Q1-Q3, kita menjaga integritas statistik data, menghasilkan data yang lebih representatif dan menghindari risiko penggantian data yang tidak valid.

### Pendekatan Robust dan Praktis Berdasarkan Jurnal Ilmiah
Studi seperti yang dipaparkan di O'Reilly Media dan ScienceDirect menyatakan bahwa penggunaan IQR dalam evaluasi dan pengendalian data sintetis membantu mempertahankan data utility sekaligus menghindari noise yang berlebihan.
Dalam paper penelitian terkait penggunaan metode probabilistik dan pembelajaran mesin untuk synthetic data (misalnya CTGAN), preservasi fitur kuartil membantu dalam replicability data dan analisis yang valid.

**Apa itu Log Transform?**
Transformasi logaritmik adalah sebuah teknik yang mengubah data asli x menjadi log(x+1). Terutama berguna untuk data yang:
- Memiliki distribusi yang sangat miring ke kanan, yaitu banyak nilai kecil dan beberapa nilai sangat besar (skewed).
- Memiliki rentang nilai yang sangat luas.
- Mempunyai nol (nilai 0), sehingga menggunakan np.log1p lebih aman daripada np.log biasa karena log(0) tidak terdefinisi.

**Tujuan Utama**
- Mengurangi skewness pada fitur, sehingga fitur menjadi lebih simetris.
- Membantu stabilisasi varians fitur, membuat datanya lebih terdistribusi normal.
- Hal ini penting karena banyak algoritma machine learning, termasuk Decision Tree, Logistic Regression, dan Linear Regression, lebih bekerja optimal dengan fitur yang terdistribusi normal.
- Membuat model lebih robust dan memperbaiki performa prediksi.

**Kenapa Gunakan np.log1p dan Bukan np.log?**
- np.log1p(x) sama dengan np.log(x + 1).
- Jika ada nilai 0 di dataset, fungsi ini tetap dapat diproses karena log(0 + 1) = 0 valid.
- Lebih numerik stabil dan aman dari error.

```python
# Log transform for skewed features
data["CO2_Reduction"] = np.log1p(data["CO2_Reduction"])
data["Investment_Cost"] = np.log1p(data["Investment_Cost"])
data
```

## Param Grid
param_grid adalah dictionary yang berisi kombinasi parameter yang ingin dicoba selama proses pencarian model terbaik menggunakan GridSearchCV.

1. 'max_depth': [3, 5, 7]
    Ini menentukan kedalaman maksimum DecisionTree. Semakin dalam pohon, semakin kompleks modelnya, tapi juga rentan overfitting.
    Nilai [3, 5, 7] artinya Grid Search akan mencoba membangun pohon dengan kedalaman maksimal 3, 5, dan 7.

2. 'min_samples_leaf': [1, 2, 4]
    Ini mengatur jumlah minimum data yang harus ada di setiap daun (leaf node). Tujuannya untuk mencegah pohon terlalu dalam dan terlalu spesifik (overfitting).
    Semakin besar nilainya, pohon akan lebih "terpotong" dan sederhana (lebih general).


- max_depth = Membatasi kedalaman maksimum pohon keputusan dan Membatasi seberapa dalam tree bisa bertumbuh, mencegah overfitting dengan menghindari pohon yang terlalu kompleks.
- min_samples_leaf = Jumlah minimum sampel yang harus ada di daun, Menentukan minimal data pada daun, membantu menghindari daun yang overfit (terlalu spesifik).
- min_samples_split = Minimum sampel yang dibutuhkan untuk membagi node menjadi cabang, Memastikan sebuah node hanya dibagi jika memiliki cukup banyak data, mencegah pembentukan cabang yang sangat kecil.
- max_leaf_nodes = Membatasi jumlah maksimum daun di dalam pohon, Membatasi total daun di pohon, kontrol kompleksitas model untuk menghindari overfitting.

Hyperparameter	Nilai	Arti dan Pengaruh
max_depth	
- 3 = Batasi kedalaman pohon maksimum hanya sampai 3 level (cukup dangkal). Pohon akan sederhana dan cenderung underfitting jika data kompleks.
- 5 = Kedalaman sampai 5 level. Ini adalah kedalaman sedang, yang biasanya bisa menangkap pola data dengan baik tanpa terlalu kompleks.
- 7 = Kedalaman sampai 7 level. Pohon lebih dalam, bisa menangkap detail lebih banyak, tapi risiko overfitting meningkat.

min_samples_leaf	
- 5 = Setiap daun harus memiliki minimal 5 sampel data. Daun tidak boleh terlalu kecil, ini membatasi kompleksitas dan mengurangi overfitting.
- 10 = Setiap daun minimal ada 10 sampel. Lebih ketat, sehingga pohon menjadi lebih sederhana dan lebih konservatif dalam membuat cabang baru.

min_samples_split	
- 4 = Sebuah node hanya dibagi (split) jika memiliki minimal 4 sampel, membatasi pembagian pada node dengan sedikit data agar tidak terlalu spesifik.
- 8 = Split hanya dilakukan jika node punya minimal 8 sampel. Ini mencegah cabang-cabang kecil yang kurang signifikan, mengurangi risiko overfit.

max_leaf_nodes	
- 5 = Batasi jumlah maksimal daun menjadi hanya 5. Pohon jadi sangat sederhana, cocok untuk data yang mudah atau untuk mengurangi overfitting signifikan.
- 10 = Maksimal ada 10 daun pohon. Lebih kompleks daripada 5 daun, memungkinkan model menangkap pola lebih detail.
- 20 = Maksimal 20 daun. Pohon bisa lebih besar dan detail, bisa meningkatkan performa tapi juga menambah risiko overfitting tergantung data.

```python
# Hyperparameter grid
param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_leaf': [5, 10],
    'min_samples_split': [4, 8],
    'max_leaf_nodes': [5, 10, 20]
}
```

**StratifiedKFold(n_splits=5, shuffle=True, random_state=42)**
- n_splits=5 Ini artinya data akan dibagi menjadi 5 fold (bagian). Proses training dan validasi akan dilakukan sebanyak 5 kali, setiap kali menggunakan 4 fold untuk training dan 1 fold untuk validasi yang bergiliran.
- shuffle=True Sebelum membagi data, data diacak terlebih dahulu untuk memastikan pembagian fold tidak bias atau berurutan. Ini baik agar representasi data menjadi lebih menyebar dan adil di setiap fold.
- random_state=42 Ini adalah seed untuk proses pengacakan (shuffling) agar hasilnya bisa direproduksi secara konsisten. Nomor ini bisa diubah sesuai keinginan, tapi menggunakan nilai tetap membantu eksperimen menjadi konsisten dan dapat direplikasi.

**Fungsi dan Tujuan StratifiedKFold**
- Stratified berarti proporsi kelas target (label) di setiap fold akan dijaga agar sama atau mirip dengan proporsi asli kelas target di dataset penuh. Ini penting untuk mencegah bias selama evaluasi, terutama pada dataset yang tidak seimbang.
- Cross-validation digunakan untuk mengevaluasi performa model dengan lebih andal dari sekadar sekali pembagian data train-test. Dengan 5 fold, kita mendapatkan rata-rata hasil yang mengurangi variansi evaluasi dari pembagian data tunggal.

**Kenapa menggunakan StratifiedKFold dan bukan KFold biasa?**
- KFold biasa membagi data tanpa memperhatikan distribusi label, sehingga pada dataset dengan kelas yang tidak seimbang, salah satu fold bisa jadi hampir tidak memiliki sampel dari kelas tertentu.
- StratifiedKFold menjaga distribusi ini supaya hasil validasi lebih representatif dan evaluasi model lebih fair.

```python
# Cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

### Apa itu GridSearchCV?
- GridSearchCV adalah teknik hyperparameter tuning otomatis di scikit-learn.
- Teknik ini mencoba semua kombinasi parameter yang diberikan dalam param_grid untuk menemukan parameter terbaik yang memberikan performa tertinggi.
- CV singkatan dari Cross-Validation, sebuah metode untuk validasi model dengan membagi data menjadi beberapa fold untuk menghindari overfitting dan mendapatkan estimasi performa yang lebih akurat.

**Parameter-parameter Penting:**
- dt: Model Decision Tree yang sudah diinisialisasi sebelumnya.
param_grid: Dictionary atau list yang berisi daftar nilai-nilai parameter yang akan diuji. Contohnya, {'max_depth': [3, 5, 7], 'min_samples_split': [4, 8]}.
- cv=cv menentukan metode validasi silang (cross-validation) yang digunakan, misalnya 5-fold atau 10-fold CV. Ini membagi data pelatihan menjadi beberapa bagian untuk melatih model dan menguji hasilnya secara bergantian, mengurangi bias evaluasi.
- scoring='accuracy': Metrik yang digunakan untuk menilai performa model pada tiap kombinasi parameter adalah akurasi.
- n_jobs=-1: Memerintahkan GridSearchCV untuk menggunakan semua core CPU yang tersedia supaya proses menjadi lebih cepat.

**Proses Kerja fit**
Fungsi fit(X_train, y_train) menjalankan grid search:
- Melakukan training dan pengujian model Decision Tree dengan semua kombinasi parameter pada data train yang sudah disediakan.
- Menggunakan 2-fold cross-validation untuk validasi internal.
- Menyimpan model terbaik yang ditemukan berdasarkan nilai akurasi tertinggi.

**Manfaat dari GridSearchCV**
- Menghindari pemilihan parameter secara manual yang mungkin kurang optimal.
- Mendapatkan model dengan parameter terbaik secara sistematis.
- Meminimalkan overfitting dengan validasi silang.
- Memaksimalkan performa prediksi model pada data baru.

### Tujuan Utama:
- Untuk menemukan kombinasi hyperparameter terbaik untuk model Decision Tree Anda.
- Menggunakan teknik cross-validation agar evaluasi parameter lebih handal dan tidak bias.
- Meningkatkan performa model dengan nilai akurasi yang optimal.
- Mempercepat proses pencarian parameter dengan penggunaan n_jobs=-1.

```python
grid_search = GridSearchCV(dt, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

**grid_search.best_estimator_**
Ini adalah model Decision Tree terbaik yang ditemukan oleh GridSearchCV, yaitu model yang sudah dilatih (fit) dengan kombinasi parameter yang memberikan hasil akurasi terbaik selama proses cross-validation.

**best_clf**
Variabel ini sekarang menyimpan model siap pakai bisa langsung menggunakannya untuk:
- Prediksi (best_clf.predict(X_test))
- Evaluasi akurasi
- Visualisasi DecisionTree

```python
# Ambil model terbaik dari hasil grid search
best_clf = grid_search.best_estimator_
```

Contoh: Dari 100 email yang diprediksi sebagai spam, berapa yang benar-benar spam?

**Recall (Sensitivitas)** adalah ukuran seberapa lengkap model menangkap semua data positif yang sebenarnya. Dari semua data yang benar-benar positif, berapa banyak yang berhasil diprediksi positif?
Contoh: Dari 100 email spam yang ada, berapa yang berhasil dikenali sebagai spam?

**F1-score** adalah nilai rata-rata harmonik dari precision dan recall yang memberikan gambaran keseimbangan antara keduanya.
Jadi, F1-score tinggi berarti model tidak hanya tepat tapi juga menangkap banyak kasus positif.

**Penjelasan Confusion Matrix**

Struktur Matriks
- Baris merupakan label sebenarnya (True Label).
- Kolom merupakan hasil prediksi model (Predicted Label).
- Setiap angka pada matriks menunjukkan jumlah sampel yang memiliki label sebenarnya pada baris tersebut dan diprediksi berada pada kolom tersebut.

Interpretasi dari matriks ini
- Baris 0: Ada 6 data dengan label sebenarnya 0 yang diprediksi dengan benar menjadi 0.
- Baris 1: Ada 2 data dengan label 1 yang diprediksi benar menjadi 1.
- Baris 2: Ada 7 data dengan label 2 yang diprediksi benar menjadi 2.
- Baris 3: Dari 5 data sebenarnya berlabel 3, 4 diprediksi dengan benar dan 1 salah diprediksi sebagai kelas 2.

Manfaat Confusion Matrix
- Memungkinkan untuk menghitung metrik evaluasi lain seperti akurasi, presisi, recall, dan F1-score.
- Membantu mengidentifikasi tipe kesalahan yang dilakukan model, misalnya apakah kelas tertentu sering salah diprediksi sebagai kelas lain.

**Penjelasan Learning Curve**

Learning curve adalah grafik yang menunjukkan performa model pada data pelatihan dan data validasi terhadap jumlah data pelatihan yang digunakan. Tujuannya adalah untuk mengukur bagaimana model belajar seiring bertambahnya data dan untuk mendeteksi masalah seperti underfitting atau overfitting.

Training accuracy menunjukkan seberapa baik model memprediksi data yang digunakan saat pelatihan.
Validation accuracy menunjukkan seberapa baik model memprediksi data baru (belum pernah dilihat sebelumnya).

**Analisis Curve yang Terlihat**

Kurva train dan validation naik secara signifikan saat ukuran data kecil hingga sedang (5-30). Ini menunjukkan model belajar dengan baik ketika jumlah data bertambah.
Akurasi train sedikit lebih tinggi dari validation, namun keduanya hampir menyatu pada nilai tinggi (sekitar 1.0). Perbedaan kecil ini menandakan model tidak mengalami overfitting yang parah. Setelah ukuran data mencapai sekitar 40-80, akurasi mulai stabil dan mencapai nilai maksimum. Ini menandakan model sudah optimal dan tidak ada manfaat signifikan menambah lebih banyak data pada titik ini.

**Apakah Model Bagus atau Overfitting?**

Ini bukan overfitting karena overfitting biasanya ditandai dengan akurasi training sangat tinggi, tetapi akurasi validasi jauh lebih rendah.
Model ini terlihat bagus (good fit) karena kinerja train dan validasi hampir sama dan baik.
Hal ini mengindikasikan model decision tree ini berhasil mempelajari pola data dengan baik dan juga mampu menggeneralisasi ke data baru.

Dalam definisinya, overfitting adalah saat model mendapatkan performa tinggi di data training tapi buruk di data validasi karena terlalu menyesuaikan noise dan detail data training.
